#Split MTET (ĐÃ CÓ CỘT DATASET -> Thuộc MTET)

```python
from datasets import DatasetDict

def split_vietai_dataset_large(vietai_dataset, dataset_name="MTet",
                               train_ratio=0.95, val_ratio=0.025, test_ratio=0.025, seed=42):
    full_data = vietai_dataset['train']
    total_size = len(full_data)

    print(f"Tổng số mẫu: {total_size:,}")

    # Thêm cột "dataset" trước khi split
    full_data = full_data.add_column("dataset", [dataset_name] * total_size)


    train_size = int(total_size * train_ratio)
    val_size = int(total_size * val_ratio)
    test_size = total_size - train_size - val_size

    print(f"Train: {train_size:,} ({train_ratio*100:.1f}%)")
    print(f"Validation: {val_size:,} ({val_ratio*100:.1f}%)")
    print(f"Test: {test_size:,} ({test_ratio*100:.1f}%)")


    shuffled = full_data.shuffle(seed=seed)


    train_data = shuffled[:train_size]
    val_data = shuffled[train_size:train_size + val_size]
    test_data = shuffled[train_size + val_size:]

    return DatasetDict({
        'train': train_data,
        'validation': val_data,
        'test': test_data
    })


vietai_splitted = split_vietai_dataset_large(vietai_dataset)

```

#Format MedEV  (ĐÃ CÓ CỘT DATASET -> Thuộc MedEV)

Row 1: English sentence 1

Row 341,001: Vietnamese translation của sentence 1

Row 2: English sentence 2

Row 341,002: Vietnamese translation của sentence 2

```python
from datasets import Dataset, DatasetDict

def process_medev_to_bilingual_proper(medev_dataset):
    """Ghép các cặp English-Vietnamese từ MedEV dataset"""
    processed_splits = {}

    for split_name, split_data in medev_dataset.items():
        total_rows = len(split_data)
        num_pairs = total_rows // 2

        print(f"  Tổng rows: {total_rows}, Số cặp: {num_pairs}")

        english_texts = []
        vietnamese_texts = []

        for i in range(num_pairs):
            en_text = split_data[i]['text']
            vi_text = split_data[i + num_pairs]['text']

            # Chắc chắn là str
            if isinstance(en_text, dict):
                en_text = str(en_text)
            if isinstance(vi_text, dict):
                vi_text = str(vi_text)

            english_texts.append(en_text)
            vietnamese_texts.append(vi_text)

        # Tạo Dataset chuẩn
        dataset_with_name = Dataset.from_dict({
            'en': english_texts,
            'vi': vietnamese_texts,
            'dataset': ["MedEV"] * len(english_texts)
        })

        processed_splits[split_name] = dataset_with_name

    return DatasetDict(processed_splits)


medev_processed = process_medev_to_bilingual_proper(medev_dataset)
```

#combined dataset

```python
from datasets import concatenate_datasets, DatasetDict

def combine_datasets(medev_processed, vietai_splitted):
    combined_splits = {}

    for split_name in ['train', 'validation', 'test']:
        if split_name in medev_processed and split_name in vietai_splitted:
            medev_split = medev_processed[split_name]
            vietai_split = vietai_splitted[split_name]

            # Chắc chắn là Dataset
            from datasets import Dataset
            if not isinstance(vietai_split, Dataset):
                vietai_split = Dataset.from_dict(vietai_split)
            if not isinstance(medev_split, Dataset):
                medev_split = Dataset.from_dict(medev_split)

            combined_dataset = concatenate_datasets([vietai_split, medev_split])
            combined_splits[split_name] = combined_dataset

            print(f"{split_name}: VietAI({len(vietai_split):,}) + "
                  f"MedEV({len(medev_split):,}) = {len(combined_dataset):,}")

    return DatasetDict(combined_splits)

combined_dataset = combine_datasets(medev_processed, vietai_splitted)
```

```python
def verify_combined_dataset(combined_dataset):


    for split_name, split_data in combined_dataset.items():
        print(f"\n{split_name.upper()}: {len(split_data)} mẫu")

        print("3 mẫu đầu tiên:")
        for i in range(min(3, len(split_data))):
            print(f"\nMẫu {i+1}:")
            print(f"  EN: {split_data[i]['en'][:100]}{'...' if len(split_data[i]['en']) > 100 else ''}")
            print(f"  VI: {split_data[i]['vi'][:100]}{'...' if len(split_data[i]['vi']) > 100 else ''}")

verify_combined_dataset(combined_dataset)
```
